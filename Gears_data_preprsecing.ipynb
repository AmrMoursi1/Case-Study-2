```python
# broken
for x in range(0,10):
    globals()[f"b30hz{x}0"] = pd.read_csv(blist[x])
b30hz70
```

```python
## add load and fault to dataset
# healthy
for x in range(0,10):
    globals()[f"h30hz{x}0"]['load']=10*x*np.ones((len(globals()[f"h30hz{x}0"].index),1))
    globals()[f"h30hz{x}0"]['fault']=np.zeros((len(globals()[f"h30hz{x}0"].index),1))
```

```python
# broken
for x in range(0,10):
    globals()[f"b30hz{x}0"]['load']=10*x*np.ones((len(globals()[f"b30hz{x}0"].index),1))
    globals()[f"b30hz{x}0"]['fault']=np.ones((len(globals()[f"b30hz{x}0"].index),1))
b30hz00
```

```python
# Combine data
# Healthy 
healthy_DF = []
healthy_df = pd.DataFrame(healthy_DF)
for x in range(0,10):
    healthy_df =  pd.concat([healthy_df,globals()[f"h30hz{x}0"]],axis=0,ignore_index=True)
healthy_df
```

```python
# Broken 
broken_DF = []
broken_df = pd.DataFrame(broken_DF)
for x in range(0,10):
    broken_df =  pd.concat([broken_df,globals()[f"b30hz{x}0"]],axis=0,ignore_index=True)
broken_df
```

```python
## Combine healthy and broken data
broken_health   = pd.concat([healthy_df,broken_df], axis =0)
broken_health
```
